## T-test

```python
def continuous_test(df, cont_var, target):
    unique_targets = df[target].unique()
    group1 = df[df[target] == unique_targets[0]][cont_var]
    group2 = df[df[target] == unique_targets[1]][cont_var]
    t_stat, p_val = stats.ttest_ind(group1, group2)
    return t_stat, p_val

def T_square_for_num(df,num_col,target):
    
    for col in num_col:
        if col == target:
            pass
        else:
            t_stat, p_val = continuous_test(df, col, target)

            print(f"######## {col} #########")

            print("T Test:")
            print(f"T Statistic: {t_stat, p_val}")
            print(f"P-Value: { p_val}")

            print(f"######## {col} #########")
            print("")
```

```python
T_square_for_num(data,cols,"Potability")
```

### T-test Results
Based on the T-test results, none of the features appear to have a statistically significant difference between the potable and not potable groups at common significance levels 

**Interpretation:**
- The P-values for all features are above 0.05, meaning we fail to reject the null hypothesis that there is no difference in the means of these features for potable and non-potable water.

- This result suggests that, individually, these features do not show a statistically significant difference between the potable and not potable classes.

T-test results provide initial insight, they might not tell the full story

## Perform multivariate tests (MANOVA) 

```python
import statsmodels.api as sm
from statsmodels.multivariate.manova import MANOVA

features = ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 
            'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']


formula = ' + '.join(features) + ' ~ Potability'


manova = MANOVA.from_formula(formula, data=data)
print(manova.mv_test())
```

The **MANOVA** results suggest that *Potability* does not have a statistically significant effect on the combination of your features.

### Interpretation of Results
**Intercept** :
- The very low p-values (Pr > F = 0.0000) for the intercept tests (Wilks' Lambda, Pillai's Trace, etc.) indicate that the overall model explains a significant amount of variability in the data.

**Potability** :
- For the Potability variable, all tests (Wilks' Lambda, Pillai's Trace, Hotelling-Lawley Trace, and Roy's Greatest Root) yield a high p-value of 0.6053. This indicates that, statistically, there is no significant difference in the multivariate means of the features between the potable and not potable groups at the common significance level (e.g., p < 0.05).

### Conclusion

The **MANOVA** test results align with the individual T-test results, suggesting that none of the features, individually or in combination, show a statistically significant difference between potable and not potable groups.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = data.drop(columns=['Potability']) 
y = data['Potability']


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)  
X_pca_2d = pca.fit_transform(X_scaled)


plt.figure(figsize=(8, 6))
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - 2D Projection')
plt.colorbar(label='Potability')
plt.show()
```

```python
from scipy.stats import ttest_ind

top_components = pd.DataFrame(X_pca_2d, columns=['PC1', 'PC2'])
top_components['potability'] = y

for pc in ['PC1', 'PC2']:
    group0 = top_components[top_components['potability'] == 0][pc]
    group1 = top_components[top_components['potability'] == 1][pc]
    t_stat, p_val = ttest_ind(group0, group1)
    print(f"Principal Component: {pc}")
    print(f"T-Statistic: {t_stat}, P-Value: {p_val}")
    print("----")
```

The T-test results on the principal components indicate that PC1 and PC2 do not show a statistically significant difference between the potability classes (0 and 1), given that the P-values for both components are well above 0.05.

**Interpretation:**
- PC1 (P-Value = 0.175): Thereâ€™s no statistically significant difference in the distribution of PC1 values between potable and non-potable classes.

- PC2 (P-Value = 0.776): Similarly, thereâ€™s no significant difference in the distribution of PC2 values between classes.


This aligns with the earlier results from the T-tests and MANOVA on the original features, reinforcing the observation that no single feature or linear combination of features appears to separate the potable from non-potable samples in a statistically significant way.

```python
from sklearn.cluster import KMeans

data['Chemical Ratios_1'] = data['Chloramines']/data['Conductivity']
data['Chemical Ratios_2'] = data['Hardness']/data['Sulfate']
data['Chemical Ratios_3'] = data['Organic_carbon']/data['Trihalomethanes']
data['Chemical Ratios_4'] = data['Organic_carbon']/data['Conductivity']
data['Chemical Ratios_5'] = data['Solids']/data['Turbidity']
data['Chemical Ratios_6'] = data['Turbidity']/data['Solids']
data['Chemical Ratios_7'] = data['ph']/data['Sulfate']
data['Chemical Ratios_8'] = data['Solids']/data['Sulfate']
data['Chemical Ratios_9'] = data['Solids']/data['Organic_carbon']
data['Chemical Ratios_10'] = data['Solids']/data['Hardness']
data['Chemical Ratios_11'] = data['Hardness']/data['Solids']
data['Chemical Ratios_12'] = data['Solids']/data['Conductivity']
data['Chemical Ratios_13'] = data['Conductivity']/data['Solids']
data['Chemical Ratios_14'] = data['Trihalomethanes']/data['Conductivity']


data['Chemical interaction_1'] = data['Chloramines']*data['Conductivity']
data['Chemical interaction_2'] = data['Hardness']*data['Sulfate']
data['Chemical interaction_3'] = data['Organic_carbon']*data['Trihalomethanes']
data['Chemical interaction_4'] = data['Organic_carbon']*data['Conductivity']
data['Chemical interaction_5'] = data['Solids']*data['Turbidity']
data['Chemical interaction_6'] = data['Turbidity']*data['Organic_carbon']
data['Chemical interaction_7'] = data['ph']*data['Sulfate']
data['Chemical interaction_8'] = data['Solids']*data['Sulfate']
data['Chemical interaction_9'] = data['Solids']*data['Organic_carbon']
data['Chemical interaction_10'] = data['Solids']*data['Hardness']
data['Chemical interaction_11'] = data['Solids']*data['Conductivity']
data['Chemical interaction_12'] = data['Trihalomethanes']*data['Conductivity']
data['Chemical interaction_13'] = data['Hardness']*data['Turbidity']
data['Chemical interaction_14'] = data['Hardness']*data['Organic_carbon']

data['ph_binned'] = pd.qcut(data['ph'], q=14, labels=False) 
data['Log_Solids'] = np.log1p(data['Solids']) 

kmeans = KMeans(n_clusters=3, random_state=42)
data['Water_Profile_Cluster'] = kmeans.fit_predict(data[cols])

pca = PCA(n_components=2)
data[['PC1', 'PC2']] = pca.fit_transform(data[cols])

tsne = TSNE(n_components=2)
data[['tsne1', 'tsne2']] = tsne.fit_transform(data[cols])

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
X = data.drop('Potability', axis=1)
y = data['Potability']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, pred))
print("Classification Report:\n", classification_report(y_test, pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, pred))
```
