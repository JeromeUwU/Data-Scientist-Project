```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import google_colab_selenium as gs
import time

```

```python
df = pd.read_csv('/content/drive/MyDrive/MAL-anime.csv')
```

```python
def add_description_to_anime(df):
  urls = df['page_url'].tolist()
  description = []

  for url in urls :
    driver = gs.Chrome()
    driver.implicitly_wait(3)
    driver.get(url)
    driver.implicitly_wait(3)

    description.append(driver.find_element(By.XPATH,'//*[@id="content"]/table/tbody/tr/td[2]/div[1]/table/tbody/tr[1]/td/p').text)
    driver.implicitly_wait(3)
    driver.quit()

  df['description'] = description

  return df
```
